```python
import pandas as pd
import matplotlib.pyplot as plt

# creating sales data
month = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [25000, 30000, 28000, 35000, 40000, 38000]

# creating dataframe
df = pd.DataFrame({
    "Month": month,
    "Sales": sales
})

# printing data
print("Sales Data")
print(df)

# calculating total sales
total = sum(sales)
print("Total Sales:", total)

# calculating average sales
average = total / len(sales)
print("Average Sales:", average)

# finding highest sales
highest = max(sales)
index = sales.index(highest)
print("Highest Sales Month:", month[index])

# line graph
plt.plot(month, sales)
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# bar chart
plt.bar(month, sales)
plt.title("Monthly Sales Bar Chart")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```
